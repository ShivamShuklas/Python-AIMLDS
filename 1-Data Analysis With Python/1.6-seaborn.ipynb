Data Visualization with Seaborn

Seaborn is a python visualization library based on Matplotlib that provides a high-level inteface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

```python
#basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
#create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#categorial plot
#bar plot
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Day vs Average Total_bill')
plt.show()
```

```python
#box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box plot of Total Bill by Day')
plt.show()
```

```python
#voilin plot
sns.violinplot(x='day', y='total_bill', data=tips) # to see distribution of data
plt.title('Violin plot of Total Bill by Day')
plt.show()
```

```python
#histograms
sns.histplot(tips['total_bill'], bins=10, kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
#kde plot
sns.kdeplot(tips['total_bill'], fill=True)
plt.title('KDE plot of Total Bill')
plt.show()
```

```python
#pairplot
sns.pairplot(tips)
plt.show()
```

```python
#heatmap
corr = tips[['total_bill', 'tip', 'size']].corr()
corr

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
#plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data=sales_df, estimator=sum) # sum of total revenue for each product category; so we dont have to do aggregation manually
plt.title('Total Sales by Product Category')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data=sales_df, estimator=sum)
plt.title('Region by Product Category')
plt.show()
```
